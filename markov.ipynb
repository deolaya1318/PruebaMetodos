```python
def likelihood(y_obs, y_model):
    chi_squared = (1.0/2.0)*sum((y_obs-y_model)**2)
    return -chi_squared
```

```python
def my_model(theta_obs, r_0, r_1, m, phi):
    return r_0 + r_1 * sin(m * theta_obs + phi)
```

```python
r0_walk = empty((0)) #this is an empty list to keep all the steps
r1_walk = empty((0))
m_walk = empty((0))
phi_walk = empty((0))
l_walk = empty((0))

r0_walk = append(r0_walk, random.random())
r1_walk = append(r1_walk, random.random())
m_walk = append(m_walk, random.random())
phi_walk = append(phi_walk, random.random())

y_init = my_model(x_obs, r0_walk[0], r1_walk[0], m_walk[0], phi_walk[0])
l_walk = append(l_walk, likelihood(y_obs, y_init))
print r0_walk
print r1_walk
print m_walk
print phi_walk
print l_walk
```

```python
n_iterations = 20000 #this is the number of iterations I want to make
for i in range(n_iterations):
    r0_prime = np.random.normal(r0_walk[i], 0.1) 
    r1_prime = np.random.normal(r1_walk[i], 0.1)
    m_prime = np.random.normal(m_walk[i], 0.1)
    phi_prime = np.random.normal(phi_walk[i], 0.1)

    y_init = my_model(x_obs, r0_walk[i], r1_walk[i], m_walk[i], phi_walk[i])
    y_prime = my_model(x_obs, r0_prime, r1_prime, m_prime, phi_prime)
    
    l_prime = likelihood(y_obs, y_prime)
    l_init = likelihood(y_obs, y_init)
    
    alpha = exp(l_prime - l_init)
    if(alpha>=1.0):
        r0_walk  = append(r0_walk, r0_prime)
        r1_walk  = append(r1_walk, r1_prime)
        m_walk  = append(m_walk, m_prime)
        phi_walk  = append(phi_walk, phi_prime)
        l_walk = append(l_walk, l_prime)
    else:
        beta = random.random()
        if(beta<=alpha):
            r0_walk  = append(r0_walk, r0_prime)
            r1_walk  = append(r1_walk, r1_prime)
            m_walk  = append(m_walk, m_prime)
            phi_walk  = append(phi_walk, phi_prime)
            l_walk = append(l_walk, l_prime)
        else:
            r0_walk = append(r0_walk, r0_walk[i])
            r1_walk = append(r1_walk, r1_walk[i])
            m_walk = append(m_walk, m_walk[i])
            phi_walk = append(phi_walk, phi_walk[i])
            l_walk = append(l_walk, l_init)
```
